```python
x = int(input('nhap so (nhap so -1 de dung)'))
if x == -1:
    print('Khong co du lieu')
else:
    max = x
    min = x
    while True:
        x = int(input('nhap so( nhap -1 de dung):'))
        if x == -1:
            break
        if x > max:
            max = x
        if x < min:
            min = x
    print('số lớn nhất là:', max)
    print('số bé nhất là:', min)
```

```python
#câu 16:
n = int(input())
chan = 0
le = 0
if n == 0:
    chan = 1
else:
    while n > 0:
        chu_so = n % 10
        if chu_so % 2 == 0:
            chan += 1
        else:
            le += 1
        n //= 10
print(f'số chữ số chẵn là: {chan}, số chữ số lẻ là: {le}') 
```

```python
n = int(input())
if n <= 0:
    print('lỗi hay nhập số nguyên dương')
else:
    while n % 3 == 0:
        n = n//3
if n == 1:
    print('n là số có dạng 3^k')
else:
    print('n không phải số có dạng n^k')
```

```python
A = int(input())
f1, f2 = 1, 1
while f1 + f2 <= A:
    f1 = f2
    f2 = f1 + f2
print(f'số fibonacci lớn nhất không vượt quá A là :{f2}')
```
