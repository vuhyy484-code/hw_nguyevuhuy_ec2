```python
a, b = map(int, input().split())
if b < a:
    print('vui lòng nhập lại dữ liệu')
else:
    tong = 0 
    for n in range(a, b + 1):
        if n < 2:
            continue
        for i in range(2, n):
            if n % i == 0:
                break
        else: 
            tong += n
    print(tong)
```

```python
#w4A10
n = int(input())
def len_collatz(x):
    cnt = 1
    while x != 1:
        if x % 2 == 0:
            x = x // 2
        else:
            x = 3 * x + 1
        cnt += 1
    return cnt
do_dai_lon_nhat = 0
so_thu_tu = 1
for i in range(1, n+1):
    L = len_collatz(i)
    if L > do_dai_lon_nhat:
        do_dai_lon_nhat = L
        so_thu_tu = i
print(so_thu_tu, do_dai_lon_nhat)

```

```python
def tong_uoc(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong
a, b = map(int, input().split())
if tong_uoc(a) == b and tong_uoc(b) == a:
    print('True')
else:
    print('False')

```

```python
#xây dựng hàm bằng cách sử dụng hàm def:( câu lệnh bên trong hàm thụt lề vào một tab)
#tham số hình thức: parameter
def tong(a, b):
    res = a+b
    return res
m,n = 10, 20
print(tong(m,n))
x, y  = 100, 200
print(tong(x, y)) #đối số = tham số chính thức = Argument
def xinchao(name1, name2, name3):
    print('Xin chao', name1, name2, name3)
xinchao("Teo",'Ty','Nam')
#trên là hàm mình xây dựng, còn ở dưới là code để chạy chg trình

x, y = map(int, input().split())
print(tong(x, y))



```
