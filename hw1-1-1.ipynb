```python
#câu 3
c = input('nhập vào một kí tự')
if len(c) > 0:
    ki_tu = c[0]
    if 'A' <= ki_tu <= 'Z':
        print('kết quả:', ki_tu.lower())
    elif 'a' <= ki_tu <= 'z':
        print('kết quả:', ki_tu.upper())
    else:
        print('đây không phải kí tự!')
else:
    print('bạn chưa nhập gì cả.')
```
